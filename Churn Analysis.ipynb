```python
import numpy as pd 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Customer Churn.csv")
```

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0").astype("float")
```

```python
df.info()
```

```python
def ctz(value):
    if value==1:
        return "Yes"
    else:
        return "No"
df["SeniorCitizen"]=df["SeniorCitizen"].apply(ctz)   
        
```

Checking how many users churn

```python
ax=sns.countplot(x= "Churn",data=df)
ax.bar_label(ax.containers[0])
plt.title("Customers Count by Churn")
plt.show()
```

seeing the percentage

```python
plt.figure(figsize=(4,4))
gb= df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'],labels=gb.index,autopct="%1.2f%%")
plt.title("Percentage of Churned customers",fontsize =20)
plt.show()
```

```python
# Create percentage crosstab
ct = pd.crosstab(df['gender'], df['Churn'], normalize='index') * 100

# Plot stacked bars
plt.figure(figsize=(5,5))
plt.bar(ct.index, ct['No'], label='No')
plt.bar(ct.index, ct['Yes'], bottom=ct['No'], label='Yes')

# Add percentage labels
for i, gender in enumerate(ct.index):
    plt.text(i, ct['No'][gender]/2, f"{ct['No'][gender]:.1f}%", ha='center', va='center')
    plt.text(i, ct['No'][gender] + ct['Yes'][gender]/2, f"{ct['Yes'][gender]:.1f}%", ha='center', va='center')

# Labels and title
plt.xlabel('Gender')
plt.ylabel('Percentage (%)')
plt.title('Churn by Gender', fontsize=16)
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

                                                   ** Analysis according Senior Citizens **

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(x="SeniorCitizen",data=df)
ax.bar_label(ax.containers[0])
plt.title("No. of Senior Citizen")
plt.show()
```

```python
# Calculate percentage crosstab
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

# Plot stacked bar chart
plt.figure(figsize=(5,5))
plt.bar(ct.index, ct['No'], label='No')
plt.bar(ct.index, ct['Yes'], bottom=ct['No'], label='Yes')

# Add percentages on bars
for i in ct.index:
    plt.text(i, ct['No'][i]/2, f"{ct['No'][i]:.1f}%", ha='center', va='center', color='black')
    plt.text(i, ct['No'][i] + ct['Yes'][i]/2, f"{ct['Yes'][i]:.1f}%", ha='center', va='center', color='black')

plt.xticks([0, 1], ['Not Senior', 'Senior'])
plt.ylabel('Percentage (%)')
plt.title('Churn by SeniorCitizen', fontsize=16)
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```

                                           **To analyze churn based on multiple services.**

```python

columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the grid size for subplots
plt.figure(figsize=(18, 14))

# Loop through each column and create a count plot
for i, col in enumerate(columns, 1):
    plt.subplot(3, 3, i)  # 3 rows, 3 columns grid
    sns.countplot(data=df, x=col, hue='Churn')
    plt.title(f'Churn by {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    
plt.tight_layout()
plt.subplots_adjust(hspace=0.3)  # Increase space between rows
plt.subplots_adjust(wspace=0.3)  # Increase space between rows


plt.show()
```

# Insights
The majority of customers who do not churn tend to have services like PhoneService, InternetService (particularly DSL), and OnlineSecurity enabled. For services like OnlineBackup, TechSupport, and StreamingTV, churn rates are noticeably higher when these services are not used or are unavailable.
